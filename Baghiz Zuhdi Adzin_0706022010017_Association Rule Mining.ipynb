```python
'''
     Aturan asosiasi (Association Rule) adalah metode data mining untuk mencari suatu hubungan yang menunjukkan kondisi di dalam 
satu set data, yang beberapa nilai atribut akan muncul secara bersamaan(Sofyan 2019). Aturan ini juga digunakan untuk menemukan
aturan asosiatif antara suatu kombinasi item atau hubungan suatu data yang digunakan untuk memprediksi suatu pola (Maulidah dan
Bachtiar 2021). Metode ini dianggap sebagai metode data mining yang tidak bergantung pada hipotesis apa pun dan dapat menemukan 
koneksi bermakna yang tersembunyi dalam kumpulan data besar.

     Association Rule ini umumnya mencari IF-Then yang sering dan berpola. Ada beberapa hal yang digunakan untuk mengukur 
sekumpulan items ini sering muncul bersamaan atau tidak (Amelia, 2021). Pentingnya aturan asosiasi dapat ditentukan oleh dua 
parameter, yaitu minimum support (presentase kombinasi item dalam database) dan minimum confidence (hubungan yang kuat antara 
item dalam aturan asosiasi), keduanya ditentukan oleh penggunanya (Fauzy, Saleh W, and Asror 2016).

     Dalam aturan ini, terdapat suatu interestingness measure (ukuran kepercayaan) yang diperoleh dari hasil pengolahan data 
dengan perhitungan tertentu digunakan untuk menentukan aturan asosiasi. Pada umumnya terdapat tiga ukuran (Rerung 2018), yaitu :
1) Support : yaitu suatu ukuran yang menunjukkan seberapa besar tingkat dominasi suatu item/itemset dari keseluruhan data yang 
diobservasi. Ukuran ini akan menentukan apakah suatu item/itemset layak untuk dicari confidence-nya (contohnya, dari seluruh 
transaksi sebuah toko, seberapa besar tingkat dominasi yang menunjukkan bahwa item A dan B dibeli secara bersamaan) dapat juga 
digunakan untuk mencari tingkat dominasi item tunggal. Secara umum dapat ditulis seperti persamaan berikut : 
Support (A→B) = Probabilitas (A→B)

2) Confidence: yaitu suatu ukuran yang menunjukkan hubungan antar dua item secara conditional (misal, seberapa sering item B 
dibeli jika orang membeli item A). Secara umum dapat ditulis seperti persamaan berikut :
Confidence (A→B) = Support (A→B) / Support A

3) Improvement/Lift: yaitu suatu ukuran yang menunjukan besarnya kemungkinan dua item dapat dibeli secara bersamaan. Lebih jelasnya 
dapat dilihat pada persamaan berikut :
Improvement (A→B) = Support (A→B) / Support(A) x Support(B)

      Ketiga ukuran ini berguna untuk membandingkan dengan threshold/batasan yang ditentukan. Batasan biasanya ditentukan dari 
min_support, min_confidence, dan min_improvement. Algoritma yang biasanya digunakan dalam Association Rule sebagai berikut :
1) Algoritma Apriori
   Algoritma apriori menggunakan pengetahuan frekuensi atribut yang telah diketahui sebelumnya untuk memproses informasi 
selanjutnya. Pada algoritma apriori menentukan kandidat yang mungkin muncul dengan cara memperhatikan minimum support dan
minimum confidence. Support (nilai penunjang) adalah persentasi kombinasi item tersebut dalam database, untuk menghitung nilai 
support pada sebuah item diperoleh dengan menggunakan persamaan (Nola Ritha, Suswaini, and Pebriadi 2021).
2) Algoritma AIS
   kumpulan item dibuat dan dihitung saat memindai data. Dalam data transaksi, algoritma AIS menentukan kumpulan item besar 
manakah yang berisi transaksi, dan kumpulan item kandidat baru dibuat dengan memperluas kumpulan item besar dengan item lain 
dalam data transaksi (Lutkevich, 2020).
3)Algoritma SETM
  Algotima ini menghasilkan kumpulan item kandidat saat memindai database, tetapi algoritme ini memperhitungkan kumpulan item 
di akhir pemindaiannya. Itemset kandidat baru dihasilkan dengan cara yang sama seperti algoritma AIS, tetapi ID transaksi dari 
transaksi yang dihasilkan disimpan dengan itemset kandidat dalam struktur data berurutan. Di akhir, jumlah dukungan kandidat 
itemset dibuat dengan menggabungkan struktur berurutan (Lutkevich, 2020).

    Penerapan implementasi dari Association Rule sebegai berikut :
1) Market-Basket Analysis
   Market basket analysis adalah suatu metode analisa atas perilaku konsumen secara spesifik dari suatu golongan / kelompok 
tertentu. Sumber data dari market basket analysis antara lain dapat bersumber dari transaksi kartu kredit, kartu lotere, kupon 
diskon, panggilan keluhan pelanggan (Gunadi and Sensuse 2012).
2) Diagnosis Medis
   Aturan asosiasi dalam diagnosis medis dapat membantu dokter mendiagnosis penyakit pasien. Diagnosis adalah proses yang sulit 
dengan banyak kesalahan potensial yang dapat menyebabkan hasil yang tidak dapat diandalkan. Penambangan aturan asosiasi 
relasional diperlukan untuk menentukan penyakit berdasarkan faktor dan gejalanya (Jena, 2022) .
3) Data Sensus
   Konsep Association Rule Mining juga digunakan dalam menangani jumlah data sensus yang sangat besar. Jika diselaraskan dengan 
benar, informasi ini dapat digunakan dalam merencanakan layanan publik dan bisnis yang efisien  (Jena, 2022).
4) Banking Service
   Digunakan oleh pengguna ritel (rekening industri uang, CD, layanan investasi, pinjaman mobil, dll.) Dengan tujuan mengenali 
pengguna yang mungkin membutuhkan layanan lain (Ginni, 2022).
       

"""
```

```python
#Import Library
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/baghizzhd/Data-Mining-association_rule_mining/main/retail_dataset.csv', sep=',')
df.head()
```

```python
# kita periksa ukuran dataset kita
df.shape
```

```python
#Data Prepocessing
items = (df['0'].unique())
items
```

```python
# Metode One-Hot Encoding digunakan untuk membantu kita melakukan ini
encoded_vals = []
for index, row in df.iterrows(): 
    labels = {}
    uncommons = list(set(items) - set(row))
    commons = list(set(items).intersection(row))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
```

```python
df2 = pd.DataFrame(encoded_vals)
```

```python
#Mari kita lihat apa yang terjadi setelah dilakukannya One-Hot Encoding
df2
```

```python
#Association Rules
'''Untuk apriori, perlu melakukannya dengan memberikan df2 dengan hot encoding yang telah dilakukan. 
Saya akan mencoba menggunakan parameter dukungan minimum untuk pemodelan ini. Untuk ini, saya menetapkan nilai min_support 
dengan nilai ambang 20%'''

freq_items = apriori(df2, min_support = 0.2, use_colnames = True, verbose = 1)
```

```python
#Dengan demikian, nilai support akan dihitung. Mari kita periksa
freq_items.head()
```

```python
#Terakhir, kita akan melihat fungsi association_rules, kita perlu menggunakan support (item frekuensi)
#Saya memilih nilai confidence minimum 60%. Dengan kata lain, ketika produk X dibeli, kita dapat mengatakan bahwa pembelian produk Y adalah 60% atau lebih.
association_rules(freq_items, metric = "confidence", min_threshold = 0.6)
```
